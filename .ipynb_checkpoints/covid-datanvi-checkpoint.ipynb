# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

# Load the Dataset

```python
dataset = pd.read_csv('owid-covid-data.csv')
dataset
```

### View column names

```python
list(dataset.columns)
```

### Only select columns that are needed for exploring

```python
df = dataset[['iso_code','continent','location', 'date', 'total_cases', 'new_cases', 'new_cases_smoothed', 
              'total_deaths','new_deaths','new_tests','total_tests','positive_rate', 'tests_units', 
              'total_vaccinations', 'population','population_density',]]
df
```

### Check each the data typ of each column to ensure proper formatting and that there are no stray elements
#### As we can see on 'tests_units' column, there is a stray data that is string and that the data type of the column is Object instead of Float. We want to eliminate this from the list.

```python
datatypeDF = pd.DataFrame(df)
dataTypeSeries = datatypeDF.dtypes
print(dataTypeSeries)
```

### Convert 'tests_units' column from Object to Float using 'to_numeric' function.
#### Then check the data types again to see if the conversion worked properly.

```python
dataset = df
```

```python
pd.set_option('mode.chained_assignment', None)
df['tests_units'] = pd.to_numeric(df['tests_units'], errors='coerce')
df
```

```python
datatypeDF = pd.DataFrame(df)
dataTypeSeries = datatypeDF.dtypes
print(dataTypeSeries)
```

## Convert Null(NaN) values to Zero (0).

```python
df.fillna(0, inplace=True)
df
```

```python
for location in df['location'].unique():
    df[df['location']==location].sort_values('total_cases').plot.line(x='date',
                                                              y='total_cases',
                                                              title=location)
```

```python

df['location'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("Count of Tips Received", labelpad=14)
plt.ylabel("Day of Week", labelpad=14)
```

```python
df.groupby(df.location, axis=0).sum()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

ax.plot(df.location, df.total_cases)
# ax.plot(df.date, df.location)
```
